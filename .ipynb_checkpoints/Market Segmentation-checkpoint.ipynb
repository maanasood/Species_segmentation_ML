```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.cluster import KMeans
```

```python
data=pd.read_csv("market_segmentation.csv")
data
```

## Plot the data

```python
plt.scatter(data['Satisfaction'],data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

```python
x=data.copy()
```

```python
kmeans=KMeans(2)
kmeans.fit(x)
```

## Resultant plot



```python
clusters=x.copy()
clusters['Cluster prediction']=kmeans.fit_predict(x)
```

```python
plt .scatter(clusters['Satisfaction'],clusters['Loyalty'],c=clusters['Cluster prediction'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

## Standarization

```python
from sklearn import preprocessing
x_scaled=preprocessing.scale(x)
x_scaled
```

## The Elbow method

```python
import warnings
warnings.filterwarnings('ignore')

wcss=[]
for i in range(1,11):
    kmeans=KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
wcss
```

```python
number_clusters=range(1,11)
plt.plot(number_clusters,wcss)
plt.xlabel('Number of clusters')
plt.ylabel('Within cluster sum of square')
plt.show()
```

### Exploring cluster solutions

### Number of clusters=2

```python
kmeans_new=KMeans(2)
kmeans_new.fit(x_scaled)
clusters_new=x.copy()
clusters_new['Clusters prediction']=kmeans_new.fit_predict(x_scaled)
clusters_new
# Thee values are not standardized 
# only the values used to predict the cluster are standardized
```

```python
plt .scatter(clusters_new['Satisfaction'],clusters_new['Loyalty'],c=clusters_new['Clusters prediction'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```
